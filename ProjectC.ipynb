```python
import pandas as pd
import numpy as np
from sklearn import model_selection
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

### Loading dataset

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python
concrete_data.isnull().sum() # checking for null values
```

### Dividing dataset into predictors and target

```python
concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] #Columns except Strength
target = concrete_data['Strength'] # Strength column
```

```python
predictors.head()
```

### Normalization of datapoints

```python
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```python
n_cols = predictors_norm.shape[1] 
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

### Building Neural Network Model¶

```python
def project_model():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

###  Train and Test the Model 50 times

```python
model=project_model()

mean_squared_error = []

for i in range(0, 50):
    X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.3)
    model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=100)
    scores = model.evaluate(X_test, y_test)
    print("\nMean Squared Error " + str(i) + ": {}".format(scores))
    mean_squared_error.append(scores)
```

```python
mse = pd.DataFrame(mean_squared_error, columns=["MSE"])
mse.head()
```

```python
mse.tail()
```

### Mean of the Mean Squared Errors

```python
np.mean(mse)
```

### Standard deviation of the Mean Squared Errors

```python
np.std(mse)
```
